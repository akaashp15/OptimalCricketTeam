```python
import pandas as pd
import numpy as np

df = pd.read_csv('Deliveries IPL 2020.csv')
df.head()
```

```python
#Calculate the runs per ball of each batsman
player_runs = df.groupby(['batsman'])['batsman_runs'].sum().sort_values()
player_balls = df.groupby(['batsman']).size()
rpb = player_runs/player_balls*100

#Calculate the number of dismissals for each batsman
outs = df['player_dismissed'].value_counts()

#Store runs per ball and number of dismissals in a dictionary
bat_dict = {}
for player in rpb.to_dict():
    try:
        bat_dict[player] = (rpb[player], outs[player])
    except:
        bat_dict[player] = (rpb[player], 0)
bat_dict
```

```python
#Calculate number of wickets taken for each bowler
wickets = df[df['player_dismissed'].notnull()]['bowler'].value_counts()

#Calculate bowler economy rate
bowler_runs = df.groupby(['bowler'])['batsman_runs'].sum().sort_values()
bowler_balls = df.groupby(['bowler']).size()
bowler_rpb = bowler_runs/bowler_balls*100


ball_dict = {}
for player in bowler_rpb.to_dict():
    try:
        ball_dict[player] = (bowler_rpb[player], wickets[player])
    except:
        ball_dict[player] = (bowler_rpb[player], 0)

print(ball_dict)
```

```python
#Calculate balls bowled per match for each bowler
balls_bowled = df.groupby(['bowler']).size()
bowl_matches = df.groupby(['bowler'])['match_id'].nunique()
balls_per_match = balls_bowled/bowl_matches
balls_per_match
```

```python
#Calculate balls batted per match for each batsman
balls_batted = df.groupby(['batsman']).size()
bat_matches = df.groupby(['batsman'])['match_id'].nunique()
batted_balls_per_match = balls_batted/bat_matches
batted_balls_per_match
```

# Optimization Model

```python
import gurobipy as gp
from gurobipy import GRB

#Create model
m = gp.Model("superteam")


#Store list of all players
players = list(set(df['bowler'].to_list() + df['batsman'].to_list()))


team = m.addVars(players, name = "team", vtype = GRB.BINARY)
m.setObjective(sum(bat_dict[p][0]*team[p] for p in bat_dict) - sum(ball_dict[p][0]*team[p] for p in ball_dict), GRB.MAXIMIZE)

#11 players on the team
m.addConstr((sum(team[p] for p in players)) == 11)

#Must have atleast 5 bowlers
m.addConstr(sum(p[1]*team[p[0]] for p in [(p, 1) if p in ball_dict else (p, 0) for p in players]) >= 5)


#Average total overs bowled by team must be atleast 20 overs
m.addConstr(sum(balls_per_match[p]*team[p] for p in ball_dict) >= 120)

#Average total overs batted by team must be atleast 20 overs
m.addConstr(sum(batted_balls_per_match[p]*team[p] for p in bat_dict) >= 120)


#Batsmen & Bowler needs to have played atleast 3 matches
#print([bat_matches[p] >= team[p] for p in bat_dict])
m.addConstrs(team[p]*3 <= bat_matches[p] for p in bat_dict)
m.addConstrs(team[p]*3 <= bowl_matches[p] for p in ball_dict)


m.optimize()

```

```python
print('\nThis is the starting lineup of the optimal team:\n')
for p in players:
    if team[p].X == 1:
        print(p)
```
